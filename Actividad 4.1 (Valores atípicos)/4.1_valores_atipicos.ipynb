```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("Detalle precios y productos fabricados 2022.xlsx")
```

```python
valores_nulos = data.isnull().sum()
valores_nulos
```

```python
data ["NOMBRE_VENDEDOR"] = data["NOMBRE_VENDEDOR"].fillna("NO REGITRADO")
```

```python
valores_nulos = data.isnull().sum()
valores_nulos
```

```python
data.info()
```

```python
cualitativas = data.select_dtypes(include=['object', 'datetime64[ns]'])
cuantitativas = data.select_dtypes(include=['int64', 'float64'])
```

```python
fig = plt.figure(figsize=(15, 8))
cuantitativas.plot(kind="box", vert=False)
plt.title("Valores atipicos del sataframe")
plt.show()
```

Desviación estandar

```python
y=cuantitativas
Limites_Superior = y.mean() + 3*y.std()
Limite_Inferior = y.mean() - 3*y.std()
print("Limite superior permitido", Limites_Superior)
print("Limite inferior pefrmitido", Limite_Inferior)
```

```python
data2=cuantitativas[(y<=Limites_Superior)&(y>=Limite_Inferior)]
data2
```

```python
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
data_clean=data2.copy()
data_clean=data_clean.fillna(round(data2.mean(),1))
data_clean
```

```python
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

Procedimiento por cuantiles

```python
y=cuantitativas

percentiles25=y.quantile(0.25) #Q1
percentiles75=y.quantile(0.75) #Q3
iqr = percentiles75 - percentiles25

Limite_Superior_iqr = percentiles75 + 1.5*iqr
Limite_Inferior_iqr = percentiles25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
data2_iqr = cuantitativas[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data2_iqr
```

```python
valores_nulos = data2_iqr.isnull().sum()
valores_nulos
```

```python
data3_iqr = data2_iqr.copy()
data3_iqr = data3_iqr.fillna(round(data2_iqr.mean(),1))
data3_iqr
```

```python
Datos_limpios = pd.concat([cualitativas, data3_iqr], axis=1)
Datos_limpios
```

```python
Datos_limpios.to_csv("Datos_limpios_4_1.csv")
```
